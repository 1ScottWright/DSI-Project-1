##### 26. Execute a hypothesis test comparing the SAT and ACT participation rates. Use $\alpha = 0.05$. Be sure to interpret your results.

```python
#H(0):  The mean participation across the states (and WDC) are the same for the SAT vs the ACT
#H(a):  The rates are different
#alpha = 0.05
import scipy.stats as stats
stats.ttest_ind(mdf['ACT_part'], mdf['SAT_part'])
```

```python
mdf['ACT_part'].mean() - mdf['SAT_part'].mean()
```

```python
#The p-value is less than alpha
#The t-test indicates that we reject the null hypothesis 
#in favor of the alternate.  Ie, it is probable that the rates are different
```

##### 27. Generate and interpret 95% confidence intervals for SAT and ACT participation rates.

```python
n=mdf.shape[0]
```

```python
CI_act_part_lower = mdf['ACT_part'].mean() - 1.96 * (mdf['ACT_part'].std())/np.sqrt(n)
CI_act_part_upper = mdf['ACT_part'].mean() + 1.96 * mdf['ACT_part'].std()/np.sqrt(n)
CI_SAT_part_lower = mdf['SAT_part'].mean() - 1.96 * mdf['SAT_part'].std()/np.sqrt(n)
CI_SAT_part_upper = mdf['SAT_part'].mean() + 1.96 * mdf['SAT_part'].std()/np.sqrt(n)
```

```python
CI_act_part_lower , CI_act_part_upper
```

```python
CI_SAT_part_lower , CI_SAT_part_upper
```

##### 29. Is it appropriate to generate correlation between SAT and ACT math scores? Why?

---

Yes, but the results are surprising.
If you dig into the data, there are some states that use the SAT or ACT as a standardized test for the entire state.  
And so, there are different populations of students that take the test.  
In mandatory states, all students of all academic abilities that the test.  
In non-mandatory states, only students who are college bound, presumably those with higher academic abilities, are taking the test.  
No state requires both, and so the correlation results are surprising.  

It might be more interesting to see the correlations between SAT and ACT parts if we exlcude high participation rates.

---
